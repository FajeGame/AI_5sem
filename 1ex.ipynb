***Модель полиномиальной регрессии (polynomial regression)*** - это обобщенная линейная модель с параметризованной степенью многочлена.

![img](https://russianblogs.com/images/753/361716febc5a15ae65cdba4394db1f69.png)

Многочлен 1-й степени:

$$ y=ax+b $$ 

Многочлен 3-й степени:
$$ y = ax^3 + bx^2 + cx + d $$

Несмотря на то, что мы можем использовать полиномиальную регрессию для моделирования нелинейных связей, она по-прежнему рассматривается как модель множественной линейной регрессии, ввиду 
линейных коэффициентов регрессии.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures  # Класс преобразователь
```

[Ссылка на документацию](https://scikit-learn.org/stable/modules/preprocessing.html#polynomial-features)

```python
# Моделируем данные
X = np.array([258.0, 270.0, 294.0, 320.0, 342.0, 368.0, 396.0, 446.0, 480.0, 586.0])[:, np.newaxis]
y = np.array([236.4, 234.4, 252.8, 298.6, 314.2, 342.2, 360.8, 368.0, 391.2, 390.8])
```

```python
X
```

```python
plt.scatter(X, y) 
```

```python
lr = LinearRegression() # Линейная регрессия
lr.fit(X, y) # Обучим линейную регрессию
```

```python
X_ = np.arange(250, 600, 10)[:, np.newaxis] # Точки для предсказания
y_lr = lr.predict(X_) # Предсказываем y для линейной регрессии
```

```python
plt.scatter(X, y) 
plt.plot(X_, y_lr) 
```

```python
pr = LinearRegression() # Полиномиальная регрессия
```

```python
quadratic = PolynomialFeatures(degree=2)
X_quad = quadratic.fit_transform(X)       # Преобразуем данные
```

```python
X_quad
```

```python
pr.fit(X_quad, y) # Обучаем полиномиальную регрессию
```

```python
y_pr = pr.predict(quadratic.fit_transform(X_)) # Предсказываем y для полиномиальной регрессии
```

```python
plt.scatter(X, y, label = 'тренировочные точки') 
plt.plot(X_, y_lr, label = 'линейная подгонка') 
plt.plot(X_, y_pr, label = 'квадратичная подгонка') 
plt.legend(loc='upper left') 
plt.show()
```

Задание 1. Оцените полученные модели с помощью коэффициента детерминации и MSE. Что можно сказать о качестве моделей?

Замечание!!! mean_squared_error принимает 1-м аргументом истинные значения y, а 2-м предсказанные (см. документацию!). 
Получается, для линейной регрессии:

```python
#
# from sklearn.metrics import mean_squared_error

# Для наших синих точек 

#y_pred = lr.predict(X)
#print('Для линейной регрессии:', mean_squared_error(y, y_pred))

# Аналогично выполните для полиномиальной модели и сравните результат

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score

# Исходные данные
x = np.array([258.0, 270.0, 294.0, 320.0, 342.0, 368.0, 396.0, 446.0, 480.0, 586.0])[:, np.newaxis]
y = np.array([236.4, 234.4, 252.8, 298.6, 314.2, 342.2, 360.8, 368.0, 391.2, 390.8])

# Линейная регрессия
lr = LinearRegression()
lr.fit(x, y)
y_pred_lr = lr.predict(x)

# Полиномиальная регрессия (степень 2)
quadratic = PolynomialFeatures(degree=2)
x_quad = quadratic.fit_transform(x)
pr = LinearRegression()
pr.fit(x_quad, y)
y_pred_pr = pr.predict(x_quad)

# Оценка моделей
mse_lr = mean_squared_error(y, y_pred_lr)
r2_lr = r2_score(y, y_pred_lr)

mse_pr = mean_squared_error(y, y_pred_pr)
r2_pr = r2_score(y, y_pred_pr)

print("=== Оценка моделей ===")
print(f"Линейная регрессия:")
print(f"  MSE: {mse_lr:.4f}")
print(f"  R²:  {r2_lr:.4f}")

print(f"\nПолиномиальная регрессия (степень 2):")
print(f"  MSE: {mse_pr:.4f}")
print(f"  R²:  {r2_pr:.4f}")

print("\nВывод:")
if r2_pr > r2_lr:
    print("Полиномиальная модель лучше объясняет дисперсию данных (выше R²).")
else:
    print("Линейная модель лучше.")
```

Задание 2. Постройте полиномиальную модель, описывающую следующую зависимость:

```python
# Новые данные
x_new = np.array([10, 12, 15, 20, 25, 30, 34, 40, 47, 54, 57])[:, np.newaxis]
y_new = np.array([80, 75, 70, 63, 65, 70, 76, 85, 90, 92, 87])

# Визуализация точек
plt.figure(figsize=(8, 5))
plt.scatter(x_new, y_new, color='blue', label='Точки')
plt.title('Исходные точки')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()

# Обучаем полиномиальную модель (степень 2)
poly_features = PolynomialFeatures(degree=2)
x_new_poly = poly_features.fit_transform(x_new)

model = LinearRegression()
model.fit(x_new_poly, y_new)

# Предсказания для визуализации
x_plot = np.linspace(8, 60, 100).reshape(-1, 1)
x_plot_poly = poly_features.transform(x_plot)
y_plot = model.predict(x_plot_poly)

# График
plt.figure(figsize=(8, 5))
plt.scatter(x_new, y_new, color='blue', label='Данные')
plt.plot(x_plot, y_plot, color='red', label='Полиномиальная подгонка (степень 2)')
plt.title('Полиномиальная регрессия для новой зависимости')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```

Задание 3. Какую степень полинома вы использовали? Попробуйте изменить степень полинома, например, на 2, 5, 15? Как меняется результат?

```python
# Функция для обучения и отображения модели с разной степенью
def plot_polynomial_regression(x, y, degrees=[2, 5, 15]):
    plt.figure(figsize=(15, 5))

    for i, degree in enumerate(degrees, 1):
        # Преобразование признаков
        poly = PolynomialFeatures(degree=degree)
        x_poly = poly.fit_transform(x)

        # Обучение
        model = LinearRegression()
        model.fit(x_poly, y)

        # Предсказания
        x_plot = np.linspace(8, 60, 100).reshape(-1, 1)
        x_plot_poly = poly.transform(x_plot)
        y_plot = model.predict(x_plot_poly)

        # Метрики
        y_pred = model.predict(x_poly)
        mse = mean_squared_error(y, y_pred)
        r2 = r2_score(y, y_pred)

        # График
        plt.subplot(1, len(degrees), i)
        plt.scatter(x, y, color='blue', label='Данные')
        plt.plot(x_plot, y_plot, color='red', label=f'Степень {degree}')
        plt.title(f'Степень {degree}\nMSE={mse:.2f}, R²={r2:.3f}')
        plt.xlabel('x')
        plt.ylabel('y')
        plt.legend()
        plt.grid(True)

    plt.tight_layout()
    plt.show()

# Вызов функции
plot_polynomial_regression(x_new, y_new, degrees=[2, 5, 15])
```
